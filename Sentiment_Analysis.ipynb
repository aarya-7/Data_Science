```python
def cleaning(tweet):
  tweet = re.sub("@[A-Za-z0-9_]+","", tweet)
  tweet = re.sub(r'\$\w*', '', tweet)
  tweet = re.sub(r'^RT[\s]+', '', tweet)
  tweet = re.sub(r'^rt[\s]+', '', tweet)
  tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)
  tweet = re.sub(r'^https[\s]+', '', tweet)
  tweet = re.sub(r'#', '', tweet)
  tweet = re.sub(r'%', '', tweet)
  tweet = re.sub(r',','',tweet)
  tweet = re.sub(r':', '', tweet)
  tweet = re.sub('[0-9]+', '', tweet)
  tweet = re.sub('\s+',' ',tweet)
  tweet = tweet.strip()
  return tweet

```

```python
data["Cleaned_data"] = data["Caption"].apply(lambda x: cleaning(x.lower()))
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stopwords = set(stopwords.words("english"))
```

```python
def clean_stopwords(tweet):
  tweet = " ".join(word for word in tweet.split() if word not in stopwords)
  return tweet
```

```python
data["stopwords_removed"] = data["Cleaned_data"].apply(clean_stopwords)
```

```python
data.rename(columns = {"LABEL":"Label"}, inplace = True)
```

```python
data.head()
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def porter(tweet):
  tweet = " ".join(ps.stem(word) for word in tweet.split() if word in tweet)
  return tweet
```

```python
data["stemmed_data"] = data["stopwords_removed"].apply(porter)
```

```python
data.head()
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def lemmatisation(tweet):
  doc = nlp(tweet)
  token = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
  return " ".join(token)
```

```python
data["Lemmatisation"] = data["stemmed_data"].apply(lemmatisation)
```

```python
data.head()
```

```python
data.rename(columns = {"Lemmatisation": "Final Tweet"}, inplace = True)
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```
